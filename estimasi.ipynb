# MODELLING DATA

## Seleksi Fitur

```python
fitur = ['year','Small_Bags','Large_Bags','Xlarge_Bags']
x = df[fitur]
y = df['AveragePrice']
x.shape, y.shape
```

## Split Data Training & Testing

```python
from sklearn.model_selection import train_test_split
x_train, X_test, y_train, y_test = train_test_split(x, y, random_state = 70)
y_test.shape
```

## membuat regresi linier

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
predik = lr.predict(X_test)
```

# EVALUTION DATA

```python
y = lr.score(X_test, y_test)
print('Akurasi Model Regresi Linier : ', y)
```

## Membuat inputan regresi linier

```python
inputan = np.array([[2017, 97601, 11720, 43417]])
prediksi = lr.predict(inputan)
print('Estimasi Jumlah Penjualan Pizza : ', prediksi)
```
